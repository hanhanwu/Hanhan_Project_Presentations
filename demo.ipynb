```
import json
import urllib
from pprint import pprint

qt = 'But with her first budget, Ms. Wynne nevertheless shifted the tone and as'
base_url = 'https://www.googleapis.com/customsearch/v1?'
query = {
        'q': qt,
        'format': 'json',
        'exactTerms': qt,
        'cx': '[your google cse cx]',
        'key': '[your google cse key]',
        'siteSearch': 'http://www.theglobeandmail.com/*'
}

url = base_url + urllib.urlencode(query)
response = urllib.urlopen(url)
data = response.read().decode('utf8')
quote = json.loads(data)

pprint(quote)
```

```
for elem in quote['items']:
  url = (elem['pagemap']['metatags'][0]['og:url'])
  print url
```

```
import re

ptn = "http://www.theglobeandmail.com/(.*?)/article(\d+)/"
m = re.match(ptn, url)
if m != None:
  print "article_id: ", m.group(2)
```
